Функция добавления записей из csv в коллекцию mongodb

```python
def csv_to_collection(filename, collection):
    data = pd.read_csv(filename)
    data_dict = data.to_dict('records')
    for list_data_dict in data_dict:
        find_doc = collection.find(list_data_dict)
        if len(list(find_doc)) != 0:
            print(list_data_dict)
            print('Такой документ существует!!!')
        else:
            try:
                collection.insert_one(list_data_dict)
                print('Документ загружен!!!')
            except pymongo.errors.DuplicateKeyError:
                print('Такой документ существует в базе данных!!!')
    return
```

Ввод вакансий в базу данных

```python
csv_to_collection('hh_vacancies_ds.csv', collection)
```

Функция, которая производит поиск и выводит на экран вакансии с заработной платой больше введённой суммы (необходимо анализировать оба поля зарплаты).

```python
def find_vacancy_gt_compensation(salary_threshold):
    get_find_vacancy = collection.find({'$or': [{'min_vacancy_compensation': {'$gt': salary_threshold}}, {'max_vacancy_compensation': {'$gt': salary_threshold}}]})
    pprint(list(get_find_vacancy))
```
