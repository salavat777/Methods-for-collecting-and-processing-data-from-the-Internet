Парсинг основных новостей с lenta.ru

```python
import requests
from pprint import pprint
import json
from lxml import html
```

```python
news_dict = {}
url = 'https://lenta.ru'
headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/109.0.0.0 Safari/537.36'
          }
req = requests.get(url=url, headers=headers)
dom = html.fromstring(req.text)
news = dom.xpath("//div[@class='card-mini__text']/span[@class='card-mini__title']")
for every_news in news:
    news_text = every_news.xpath("./text()")[0]
    try:
        news_date = every_news.xpath("../div/time[@class='card-mini__date']/text()")[0]
    except IndexError:
        news_date = ''
    news_link = url + every_news.xpath("../../@href")[0]
    news_dict[news_text] = {
        'date': news_date,
        'link': news_link
    }
news_dict
```
