```python
import requests
from pprint import pprint
import json
```

```python
user = 'salavat777'
url = 'https://api.github.com/users/'+user+'/repos'
headers = {
    'Accept': 'application/vnd.github+jso'
          }
req = requests.get(url=url, headers=headers)
data = req.json()
with open('data.json', 'w') as f:
    json.dump(data, f)
with open('data.json') as f:
    data2 = json.load(f)
pprint(data2)
```
